```python
# Importar librerías para tratamiento de datos
# -----------------------------------------------------------------------
import pandas as pd  # Pandas es una poderosa librería para manipulación y análisis de datos en Python.

# Importar librerías para automatización de navegadores web con Selenium
# -----------------------------------------------------------------------
from selenium import webdriver  # Selenium es una herramienta para automatizar la interacción con navegadores web.
from selenium.webdriver.common.keys import Keys  # Keys es útil para simular eventos de teclado en Selenium.
from selenium.webdriver.support.ui import Select  # Select se utiliza para interactuar con elementos <select> en páginas web.

# Importar librerías para web scraping y manipulación de datos
# -----------------------------------------------------------------------
from bs4 import BeautifulSoup
import requests

# Importar librerías para pausar la ejecución
# -----------------------------------------------------------------------
from time import sleep  # Sleep se utiliza para pausar la ejecución del programa por un número de segundos.

# Configuraciones
# -----------------------------------------------------------------------
pd.set_option('display.max_columns', None)  # Establece una opción de Pandas para mostrar todas las columnas de un DataFrame.
```

```python
## Orden BBDD Procuctores Locales
# 0. nombre / marca
# 1. categoria (lacteos, huevos, bebidas, bebidas alcoholicas, fruta, verdura)
# 2. direccion
# 3. localidad
# 4. provincia
# 5. ccaa
# 6. telefono
# 7. mail
# 8. web -
# 9. productos
# 10. tipo (oveja, txakoli)
# 11. sello de calidad
# 12. url_imagen 
# 13. descripcion
# 14. facebook 
# 15. twitter 
# 16. google 
# 17. instagram 
# 18. Linkedin 
# 19. venta_online (true / False) 

# -->FALTA VENTA ONLINE, LINKEDIN Y TIPO (OVEJA) EN IDIAZABAL
```

```python
# Inicializa el navegador Firefox
driver = webdriver.Firefox()

driver.maximize_window()
driver.get("https://txakolidealava.eus/bodegas/")


lista_datos =[]
s = 2
while True:
    try:
        sleep(1)
        # recoger datos de cada uno de los productores (Cambia la posición en la web si es una sección par o impar)

        if s%2==0:
            i = 1
        else:
            i = 2    

        datos = driver.find_element("xpath", f"/html/body/div[1]/div[3]/article/div/div/div/div/div/div[1]/section[{s}]/div[2]/div[{i}]/div").text
        # web = driver.find_element("xpath", f"/html/body/div[1]/div[3]/article/div/div/div/div/div/div[1]/section[{s}]/div[2]/div[{i}]/div/div[5]/div/div/a").get_attribute('href')

        ## Sacar redes sociales con BeautifulSoup
        #----------------------------------------

        # definimos la url de la página de la vamos a sacar datos
        pagina = driver.find_element("xpath", f"/html/body/div[1]/div[3]/article/div/div/div/div/div/div[1]/section[{s}]/div[2]/div[{i}]/div/div[5]/div/div/a").get_attribute('href')
        
        # Request a la página de la que queremos sacar la información
        respuesta = requests.get(pagina)

        # vemos si todo ha ido bien
        print("La respuesta de la petición es:", respuesta.status_code)

        # Preparar la sopa
        sopa_txakoli = BeautifulSoup(respuesta.content, 'html.parser')

        #buscar enlaces a redes sociales en la sopa
        try:
            facebook = sopa_txakoli.select_one("a[href*=facebook]").get('href')
        except:
            facebook = 'No disponible'
        try:
            twitter = sopa_txakoli.select_one("a[href*=twitter]").get('href')
        except:
            twitter = 'No disponible'
        try:    
            instagram = sopa_txakoli.select_one("a[href*=instagram]").get('href')
        except:
            instagram = 'No disponible'
        try:    
            linkedin = sopa_txakoli.select_one("a[href*=linkedin]").get('href')
        except:
            linkedin = 'No disponible'
        
        # añadir los datos  a la lista
        lista_productor = [datos, facebook, twitter, instagram, linkedin]
        lista_datos.append(lista_productor)

        print(lista_productor)
        s += 1
        driver.back()     

    except:
        break

# Cierra el navegador
driver.close()
```

```python
datos_txakoli_alava = []
direccion = 'No disponible'
localidad = 'No disponible'
telefono = 'No disponible'
mail = 'No disponible'
web = 'No disponible'
google = 'No disponible'
venta_online = False
url_imagen = 'No disponible'

# Separar y ordenar los elementos de los diferentes datos de cada uno de los productores
for elemento in lista_datos:
    # Añadir datos de extracción sinmodificar
    lista = elemento[0].split('\n')
    nombre = lista[1]
    categoria = 'Bebidas Alcohólicas'
    facebook = elemento[-4]
    twitter = elemento[-3]
    instagram = elemento[-2]
    linkedin = elemento[-1]
    
    # Añadir imagen y venta online
    if 'beldio' in nombre.lower():
        url_imagen = 'http://www.beldui.com//images/Logo-BELDUI---RGB.png'
        venta_online = True
    if 'tantaka' in nombre.lower():
        url_imagen = 'http://www.tantaka.eus//img/logo.png'
        venta_online = False
    if 'artomaña' in nombre.lower():
        url_imagen = 'https://artomanatxakolina.eus/wp-content/uploads/2020/11/LogoSiltxakolina.png'
        venta_online = True
    if 'astobiza' in nombre.lower():
        url_imagen = 'https://www.astobiza.es/wp-content/uploads/2021/01/astobiza-logo-150x91.png'
        venta_online = True
    if 'gara' in nombre.lower():
        url_imagen = 'https://www.batgara.com/nueva/wp-content/uploads/2019/04/LOGO-BAT-GARA.png'            
     
    if 'Teléfono de contacto' not in lista[2]:
        descripcion = lista[2]
    else:
        descripcion = 'No disponible'
    
    for dato in lista:

        if  'Teléfono de contacto' in dato:
            telefono =  dato.split(':')[1].strip()

        if  'Email' in dato:
            mail =  dato.split(':')[1].strip()

        if  'Dirección' in dato:
            direccion =  dato.split(':')[1][:-7].strip()
            localidad =  direccion.replace(',', '').split()[-1]

        if  'Sitio web' in dato:
            web =  dato.split(':')[1].strip().split()[0]
            
    # Añadimos los datos que son iguales para todos los productores de Txakolí de Álava       
    provincia =  'Álava'
    ccaa = 'País Vasco'
    productos = 'Vino'
    tipo = 'Txakoli'
    sello_calidad = 'DOP Txakolí de Álava'
    lista_productor = [nombre, categoria, direccion, localidad, provincia, ccaa, telefono, mail, web,  productos, tipo, sello_calidad, url_imagen, descripcion, facebook, twitter, google, instagram, linkedin, venta_online]
    datos_txakoli_alava.append(lista_productor)

```

```python
# Mostramos los datos en un dataframe
df_txakoli_alava = pd.DataFrame(datos_txakoli_alava)
df_txakoli_alava
```
