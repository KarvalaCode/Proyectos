```python
lineas_pag1 = pymupdf_text[0].split('\n')
```

```python
dict_lineas = {}
lista_lineas = []
contador = 0
for linea in lineas_pag1:
    linea = linea.strip()
    if linea != '' and linea != ' ':
        lista_lineas.append(linea)
        if ': ' in linea:
            lista_linea = linea.split(':', 1)
            dict_lineas[lista_linea[0]] = lista_linea[1].strip() 
        else:
            contador += 1
            dict_lineas[f'otro{contador}'] = linea

dict_lineas         
```

```python
lista_lineas
```

```python
for key in dict_lineas.keys():
    if 'importe' in key.lower():
        diccionario_final["importe_factura"] = dict_lineas[key]
    if 'nº' in key.lower():
        diccionario_final["número_factura"] = dict_lineas[key]
    if 'periodo' in key.lower():
        fechas_consumo = dict_lineas[key].split('a')
        diccionario_final["inicio_periodo"] = fechas_consumo[0].strip()
        diccionario_final["fin_periodo"] = fechas_consumo[-1].strip()
    if 'fecha' in key.lower():
        diccionario_final["fecha_cargo"] = dict_lineas[key]     
    if 'potencia' in key.lower():
        diccionario_final["potencia_contratada"] = dict_lineas[key]     
    if 'titular' in key.lower():
        diccionario_final["nombre_cliente"] = dict_lineas[key]     
    if 'nif' in key.lower():
        diccionario_final["dni_cliente"] = dict_lineas[key]     
    if 'nif' in key.lower():
        diccionario_final["dni_cliente"] = dict_lineas[key]     
    if 'domicilio social' in key.lower():
        diccionario_final["provincia_comercializadora"] = dict_lineas[key].split('-')[-1].strip()   

for key in dict_lineas.keys():
    if 'cif' in key.lower():
        diccionario_final["cif_comercializadora"] = dict_lineas[key]  
        break 

for key in dict_lineas.keys():
    
        break 
```

```python
# File to read
filename = "pdfs/factura_915.pdf"

credentials = Credentials.service_account_credentials_builder().from_file(filename).build()

execution_context = ExecutionContext.create(credentials)
extract_pdf_operation = ExtractPDFOperation.create_new()

#Set operation input from a source file.
source = FileRef.create_from_local_file(filename)
extract_pdf_operation.set_input(source)

# Build ExtractPDF options and set them into the operation
extract_pdf_options: ExtractPDFOptions = ExtractPDFOptions.builder().with_elements_to_extract([ExtractElementType.TEXT, ExtractElementType.TABLES]).with_elements_to_extract_renditions([ExtractRenditionsElementType.TABLES,ExtractRenditionsElementType.FIGURES]).build()
  

extract_pdf_operation.set_options(extract_pdf_options)

#Execute the operation.
result = extract_pdf_operation.execute(execution_context)

result
```
